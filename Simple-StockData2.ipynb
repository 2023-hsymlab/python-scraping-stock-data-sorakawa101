```python
# SoftBank Stock data from Stooq.com
# reference https://seanmemo.com/234/

from pandas_datareader import data

def make_df():
    start = 'JAN-01-2013'
    end = 'April-19-2023'
    # df = data.DataReader('9984.JP', 'stooq', start, end)
    df = data.StooqDailyReader('9984.JP', start, end).read()
    return df
```

```python
make_df()
```

```python
import numpy as np
import plotly.graph_objects as go

def chart_option(df):
    x = np.arange(df.shape[0])

    interval = 20
    vals = np.arange(df.shape[0],step=interval)
    labels = list(df[::interval].index)

    return x, vals, labels
```

```python
x, vals, labels = chart_option(df)
```

```python
def plot_candle_chart(x, vals, labels):

    fig = go.Figure(
            data=[go.Candlestick(
                    x=x,
                    open=df['Open'],
                    high=df['High'],
                    low=df['Low'],
                    close=df['Close']),
                    hovertext= ['date:{}<br>open:{}<br>high:{}<br>low:{}<br>close:{}'
                                .format(df.loc[i,'date'],df.loc[i,'open'],df.loc[i,'high'],df.loc[i,'low'],df.loc[i,'close']) for i in range(len(df))] ,
                ],
            layout = go.Layout(
                    xaxis = dict(
                        tickvals = vals,
                        ticktext = labels,
                        tickangle=-45
                    ),
            )
    )

    fig.show()
```

```python
plot_candle_chart(x, vals, labels)
```
